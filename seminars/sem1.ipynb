```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
me = cv2.imread("../data/me.jpg", cv2.IMREAD_COLOR_RGB)
milkyway = cv2.imread("../data/milky-way.jpg", cv2.IMREAD_COLOR_RGB)
house = cv2.imread("../data/house.jpeg", cv2.IMREAD_COLOR_RGB)
blocks = cv2.imread("../data/blocks.webp", cv2.IMREAD_COLOR_RGB)
```

```python
kernel = np.ones((5,5),np.uint8)
plt.imshow(cv2.erode(house, kernel))
```

```python
kernel = np.ones((10,10),np.uint8)
plt.imshow(cv2.erode(house, kernel))
```

```python
kernel = np.ones((5,5),np.uint8)
plt.imshow(cv2.dilate(house, kernel))
```

```python
kernel = np.ones((5,5),np.uint8)
plt.imshow(cv2.morphologyEx(milkyway, cv2.MORPH_OPEN, kernel))
```

```python
kernel = np.ones((5,5),np.uint8)
plt.imshow(cv2.morphologyEx(milkyway, cv2.MORPH_CLOSE, kernel))
```

```python
kernel = np.ones((5,5),np.uint8)
plt.imshow(cv2.morphologyEx(milkyway, cv2.MORPH_GRADIENT, kernel))
```

```python
kernel = np.ones((5,5),np.uint8)
plt.imshow(cv2.morphologyEx(house, cv2.MORPH_GRADIENT, kernel))
```

```python
kernel = np.ones((5,5),np.uint8)
plt.imshow(cv2.morphologyEx(milkyway, cv2.MORPH_TOPHAT, kernel))
```

```python
kernel = np.ones((5,5),np.uint8)
plt.imshow(cv2.morphologyEx(house, cv2.MORPH_BLACKHAT, kernel))
```

```python
plt.imshow(cv2.Sobel(house, 3, dx=0, dy=1))
```

```python
plt.imshow(cv2.Sobel(house, 3, dx=1, dy=0))
```

```python
edges = cv2.Canny(house, 100, 200)
plt.imshow(edges)
```

```python
blocks = cv2.imread("../data/blocks.webp", cv2.IMREAD_COLOR_RGB)
gray = cv2.cvtColor(blocks, cv2.COLOR_BGR2GRAY)
dst = cv2.cornerHarris(gray, 2, 3, 0.04)
dst = cv2.dilate(dst,None)
blocks[dst>0.08*dst.max()]=[0,0,255]
plt.imshow(blocks)
```
